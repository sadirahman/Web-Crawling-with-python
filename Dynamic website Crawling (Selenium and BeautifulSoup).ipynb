```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd


#URL of the page
url= 'https://marketplace.atlassian.com/search?category=Time%20tracking&product=jira'

#I have used my crome driver path, You have to use yous to run it
driver_path = "C:/Users/Sun of sadi/PycharmProjects/Web_Crawler/chromedriver.exe"

driver = webdriver.Chrome(driver_path)
driver.get(url)

time.sleep(15)

#Parsing into html using html.parser
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

for i in range(0,10):
    btn_element = driver.find_element_by_xpath('/html/body/section/div/div[3]/div/div/div/div/div/div[2]/div/div/div/div[2]/div/div/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[2]/div/div/div/div/div[2]/div/div/div/div/div/button')
    btn_element.click()

    time.sleep(15)

    # Parsing into html using html.parser
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")


#This is for getting name of the products:
titles = []
all_titles = soup.find_all('div',{"class":"css-186q604-HeadingContainer"})
for title in all_titles:
    name = title.span.text
    titles.append(name)


#This is for getting number of votes and number of installs:

number_of_people_rated = []
num_of_installs = []
all_ratting_data = soup.find_all("div",{"class":"css-1ug7nz5-Footer"})

for all in all_ratting_data:
    data = all.text.split(')')
    if len(data)==0 or data[0]=="New":
        data.insert(0,'N/A')
        data.insert(0,'N/A')
    else:
        if len(data)==1:
            data[0] = str(data[0]).replace('(','')
            data.insert(0,'N/A')
        if len(data)==2:
            if len(data[1]) == 0:
                data[0] = str(data[0]).replace('(', '')
                data.insert(1,'N/A')
            else:
                data[0] = str(data[0]).replace('(', '')
                stf = data[1].split(" ")
                if stf[0] == "Staff":
                    data[1] = "N/A"
                else:
                    data[1] = stf[0]

    number_of_people_rated.append(data[0])
    num_of_installs.append(data[1])

df = pd.DataFrame({'Product Name' : titles, 'Number of Install':num_of_installs,'Get rating':number_of_people_rated})
df.to_csv("Atlassian_Marketplace_data.csv")
print(df)
```
