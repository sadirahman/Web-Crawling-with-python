```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

#URL of the page
url= 'https://marketplace.atlassian.com/search?category=Time%20tracking&product=jira'

#I have used my crome driver path, You have to use yous to run it
driver_path = "C:/Users/Sun of sadi/PycharmProjects/Web_Crawler/chromedriver.exe"

driver = webdriver.Chrome(driver_path)
driver.get(url)



#Putting into sleep for giving time to load the page successfully
time.sleep(20)

#Parsing into html using html.parser
html = driver.page_source

soup = BeautifulSoup(html, "html.parser")

#This is for getting number of installs
number_of_installs = []
all_installs = soup.find_all("div",{"class":"css-ke6n3l-FooterRightContainer"})
for installs in all_installs:
    data = str(installs.text)
    in_data=data.split()
    if in_data[0] =="Staff":
        number_of_installs.append("N/A")
    else:
        number_of_installs.append(in_data[0])
       


#This is for getting number of votes
number_of_people_rated = []
all_rattings = soup.find_all("span",{"class":"css-10hmxqp-RatingsCount"})

for tag in all_rattings:
    num = tag.text
    in_num  = num.split()
    for i in in_num:
        number_of_people_rated.append(i)
    

titles = []
all_titles = soup.find_all('div',{"class":"css-186q604-HeadingContainer"})
for title in all_titles:
    name = title.span.text
    titles.append(name)
    


data_in = pd.DataFrame({'Titles':titles,'Install':number_of_installs,'Vote':number_of_people_rated})
data_in.to_csv (r'C:/Users/Sun of sadi/PycharmProjects/Web_Crawler/Software_products_list.csv', index = False, header=True)
print(data_in)
```
